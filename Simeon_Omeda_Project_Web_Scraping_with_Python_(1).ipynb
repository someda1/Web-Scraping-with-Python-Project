## <font color='#2F4F4F'>Prerequisites</font>

```
# We first import the required libraries
# ---
#
import pandas as pd             # library for data manupation
import requests                 # library for fetching a web page 
from bs4 import BeautifulSoup   # library for extrating contents from a webpage 
```

## <font color='#2F4F4F'>Step 1: Obtaining our Data</font>

```
# PigiaMe: https://www.pigiame.co.ke/it-software-jobs
# ---
#
pigia_me = requests.get('https://www.pigiame.co.ke/it-software-jobs')
pigia_me
```

```
# MyJobMag: https://www.myjobmag.co.ke/jobs-by-field/information-technology
# ---
#
my_job_mag = requests.get('https://www.myjobmag.co.ke/jobs-by-field/information-technology')
my_job_mag
```

```
# KenyanJob: https://www.kenyajob.com/job-vacancies-search-kenya?f%5B0%5D=im_field_offre_secteur%3A133
# ---
#
kenyan_job = requests.get('https://www.kenyajob.com/job-vacancies-search-kenya?f%5B0%5D=im_field_offre_secteur%3A133')
kenyan_job
```

## <font color='#2F4F4F'>Step 2: Parsing</font>

```
# Parsing our document: pigia_me
# ---
# 
pigia_me_soup = BeautifulSoup(pigia_me.text, "html.parser")
```

```
print(pigia_me.text[0:1600])
```

```
# Parsing our document: my_job_mag
# ---
#  
my_job_mag_soup = BeautifulSoup(my_job_mag.text, "html.parser")
```

```
print(my_job_mag.text[0:1600])
```

```
# Parsing our document: kenyan_job
# ---
# 
kenyan_job_soup = BeautifulSoup(kenyan_job.text, "html.parser")
```

```
print(kenyan_job.text[0:1600])
```

## <font color='#2F4F4F'>Step 3: Extracting Required Elements</font>

```
# 1. Extracting job titles and links: pigia me
# We then create empty lists that we will use to store
pigia_me_link_content = []
pigia_me_link_url = []

pigia_me_results = pigia_me_soup.find_all('div', class_ = 'listing-card__header__title')
pigia_me_links = pigia_me_soup.find_all('a', {'class': "listing-card__inner"})

for xp in pigia_me_results:
   
    # Getting our text from each tag
    pigia_me_text = xp.get_text()
    pigia_me_link_content.append(pigia_me_text)
  

for links in pigia_me_links:
    pigia_me_link_url.append(links.get('href'))
```

```
# Previewing our link_content list by checking first 10 items
# ---
#
pigia_me_link_content[0:10]
```

```
# filter unwanted text ['\n']
pigia_me_job_titles = list()
for x in pigia_me_link_content:
    pigia_me_job_titles.append(x.replace("\n", ""))

pigia_me_job_titles
```

```
# Previewing our links_content list by checking first 10 items
# ---
#
pigia_me_link_url[0:10]
```

```
# 2. Extracting job titles: my_job_mag_
# We then create empty lists that we will use to store
my_job_mag_link_content = []
my_job_mag_link_url = []

my_job_mag_results = my_job_mag_soup.find_all('li', class_ = "mag-b")

for mj in my_job_mag_results:
   
    # Getting our text from each tag
    my_job_mag_text = mj.get_text()
    my_job_mag_link_content.append(my_job_mag_text)

for job in my_job_mag_results:
    if job.h2:
    
      link ='https://www.myjobmag.co.ke' + job.h2.a['href']
      
      my_job_mag_link_url.append(link)
```

```
# Previewing our link_content list by checking first 10 items
# ---
#
my_job_mag_link_content[0:10]
```

```
# filter unwanted text ['\n']
my_job_mag_job_titles = list()
for x in my_job_mag_link_content:
    my_job_mag_job_titles.append(x.replace("\n", ""))

my_job_mag_job_titles
```

```
# Previewing our links_content list by checking first 10 items
# ---
#
my_job_mag_link_url[0:10]
```

```
# 3. Extracting job titles: kenya_job_
# We then create empty lists that we will use to store
kj_url = 'https://www.kenyajob.com'

kenyan_job_link_content = []
kenyan_job_link_url = []

kenyan_job_results = kenyan_job_soup.find_all('div', {'class': "col-lg-5 col-md-5 col-sm-5 col-xs-12 job-title"})

for t in kenyan_job_results:
    kenyan_job_link_content.append(t.a.get_text())
    kenyan_job_link_url.append(kj_url + t.a.get('href'))
```

```
# Previewing our link_content list by checking first 10 items
# ---
#
kenyan_job_link_content[0:10]
```

```
# Previewing our links_content list by checking first 10 items
# ---
#
kenyan_job_link_url[0:10]
```

## <font color='#2F4F4F'>Step 4: Saving our Data</font>

Pigia Me

```
# Saving the scraped contents in a dataframe and preview our data
#
pigia_me_df = pd.DataFrame({"job_title": pigia_me_job_titles, "links": pigia_me_link_url})
pigia_me_df.head()
```

My Job Mag

```
# Saving the scraped contents in a dataframe and preview our data
#
my_job_mag_df = pd.DataFrame({"job_title": my_job_mag_job_titles, "links": my_job_mag_link_url})
my_job_mag_df.head()
```

Kenya Job

```
# Saving the scraped contents in a dataframe and preview our data
#
kenyan_job_df = pd.DataFrame({"job_title": kenyan_job_link_content, "links": kenyan_job_link_url})
kenyan_job_df.head()
```

CONCATENATE THE DFs

```
# Saving the scraped contents in a dataframe and preview our data
# ---
#
jobs_df = pd.concat([pigia_me_df, my_job_mag_df, kenyan_job_df], ignore_index=True)
jobs_df.head(30)
```
